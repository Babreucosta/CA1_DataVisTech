
• Existe uma correlação entre “users_rated” e “max_playtime”?


```python
# Calculate the Pearson correlation coefficient
correlation = board_games['users_rated'].corr(board_games['max_playtime'])

# Display the correlation coefficient
correlation
```

• What is the distribution of game categories? (You cannot use a bar chart)

```python
# Calculate the distribution of game categories
category_distribution = board_games['category'].value_counts()

# Create a pie chart
plt.figure(figsize=(10, 10))
plt.pie(category_distribution, labels=category_distribution.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Game Categories')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate the distribution of game categories
category_distribution = board_games['category'].value_counts()

# Select the top N categories to display (adjust N as needed)
top_n_categories = 10  # You can change this number to show more or fewer categories

# Plot a horizontal bar chart for the top N categories
plt.figure(figsize=(10, 6))
category_distribution.head(top_n_categories).plot(kind='barh', color='skyblue')
plt.title('Distribution of Game Categories')
plt.xlabel('Frequency')
plt.ylabel('Category')
plt.gca().invert_yaxis()  # Invert the y-axis for better readability
plt.show()
```

```python
import matplotlib.pyplot as plt

# Get the top 10 categories for better visualization (optional)
top_categories = category_distribution.head(10)

# Create a horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(top_categories.index, top_categories.values, color='skyblue')
plt.xlabel('Frequency')
plt.ylabel('Category')
plt.title('Distribution of Game Categories')
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Concatenate all categories into a single string
all_categories = ''.join(board_games['category'].dropna())

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_categories)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import squarify
import matplotlib.pyplot as plt

# Count the frequency of each game category
category_counts = board_games['category'].value_counts()

# Create a color palette for the treemap
colors = plt.cm.Paired(range(len(category_counts)))

# Create a squarify treemap
plt.figure(figsize=(10, 6))
squarify.plot(sizes=category_counts.values, label=category_counts.index, color=colors, alpha=0.7)
plt.axis('off')
plt.title('Distribution of Game Categories (Treemap)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a violin plot to visualize the distribution of game categories
plt.figure(figsize=(10, 6))
sns.violinplot(x=board_games['category'], inner='quartile', palette='pastel')
plt.xticks(rotation=90)
plt.title('Distribution of Game Categories (Violin Plot)')
plt.xlabel('Game Category')
plt.ylabel('Frequency')
plt.show()
```

• Do older games (1992 and earlier) have a higher MEDIAN “average rating” than newer games (after 1992)?

#### Split the Data:
Divide the dataset into two groups: games released in 1992 and earlier, and games released after 1992.

```python
older_games = board_games[board_games['year_published'] <= 1992]
newer_games = board_games[board_games['year_published'] > 1992]
```

#### Calculate Median Average Ratings:

```python
median_rating_older = older_games['average_rating'].median()
median_rating_newer = newer_games['average_rating'].median()
```

```python
# Compare the Medians
median_rating_older, median_rating_newer
```

• What are the five most common “mechanics” in the dataset?

```python
# Count the frequency of each game mechanic
top_mechanics = board_games['mechanic'].value_counts().head(5)

# Display the top 5 mechanics
top_mechanics
```

```python
import matplotlib.pyplot as plt

# Get the top 5 mechanics
top_mechanics = board_games['mechanic'].value_counts().head(5)

# Create a horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(top_mechanics.index, top_mechanics.values, color='skyblue')
plt.xlabel('Frequency')
plt.ylabel('Mechanic')
plt.title('Top 5 Most Common Mechanics')
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.show()
```
